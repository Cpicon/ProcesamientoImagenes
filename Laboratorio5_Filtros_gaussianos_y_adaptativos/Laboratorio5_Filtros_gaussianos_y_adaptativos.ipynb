```python
def make_gaussian_kernel2(size, sigma):
    distance = size//2
    x, y = np.meshgrid(range(-distance,distance+1), range(-distance,distance+1))
    mu = 0.0
    g1 = np.exp( - (((x - mu)**2+(y - mu)**2) / (2 * sigma**2) ))
    kernel_2d = g1 / (2 * np.pi * sigma ** 2) # unit integral
    kernel_2d = kernel_2d + (1 - np.sum(kernel_2d))/(size*size)
    print("El kernel suma un total de : ",np.sum(kernel_2d))
    return kernel_2d
    
size = 7
sigma = 1.0
g2 = make_gaussian_kernel2(size, sigma)
```

## Punto 3. Filtro mediano adaptativo

```python
def MyAdaptMedian_201619749_201126292(image, window_size, max_window_size):
    # copy the image
    img_copy = np.copy(image)
    # get widht of pad
    pad = max_window_size//2
    #store window_size
    window_size_original = window_size
    #make padding with zeros
    padding_img = np.ones(shape=(image.shape[0]+2*pad, image.shape[1]+2*pad))
    padding_img[pad:image.shape[0]+pad,pad:image.shape[1]+pad] = image
    #iterate over rows in original image
    for row in range(pad,image.shape[0]+pad):
        # create a window slice
        #iterate over columns in original image
        for column in range(pad,image.shape[1]+pad):
            # return to original size of window
            window_size = window_size_original
            # resize the window to first original size
            inicio = (window_size+2)//window_size
            #get the window over original image
            ventana = padding_img[row-inicio:row+inicio+1, column-inicio:column+inicio+1]
            while True:
                # flatten the window
                ventana = ventana.ravel()
                # organice the window in ascending way
                ventana_sort = np.sort(ventana)
                #get minimun
                Zmin = ventana_sort[0]
                #get maximun
                Zmax = ventana_sort[-1]
                # get median
                Zmed = np.median(ventana_sort)
                #get central pixel value
                Zxy = image[row-pad,column-pad]
    
                A1 =  Zmed - Zmin
                A2 = Zmed - Zmax
                if (A1>0) and (A2<0):
                    #Si 𝑍𝑚𝑖𝑛 < 𝑍𝑚𝑒𝑑 < 𝑍𝑚𝑎𝑥 significa que 𝑍𝑚𝑒𝑑 NO es un impulso (por la hipotesis), solo entonces se ejecuta
                    B1 = Zxy - Zmin
                    B2 = Zxy -Zmax
                    if (B1>0) and (B2<0):
                        # overwrite copy image by the central pixel 
                        img_copy[row-pad,column-pad]=Zxy
                        break
                    else:
                        # overwrite copy image by the median pixel 
                        img_copy[row-pad,column-pad]=Zmed
                        break
                else:
                    #Si 𝑍𝑚𝑒𝑑 = 𝑍𝑚𝑖𝑛 o 𝑍𝑚𝑒𝑑 = 𝑍𝑚𝑎𝑥 significa que 𝑍𝑚𝑒𝑑 es un impulso, entonces se incrementa el tamaño de la ventana y repite la Etapa A.
                    if window_size>=max_window_size:
                        # overwrite copy image by the median pixel
                        img_copy[row-pad,column-pad]=Zmed
                        # the window_size cant be bigger than max_window_size
                        window_size = max_window_size
                        # make maximun window
                        ventana = padding_img[row-pad:row+pad+1, column-pad:column+pad+1]
                        break
                    else:
                        #increase window size in two
                        window_size += 2
                        # make new window
                        inicio = window_size//2
                        ventana = padding_img[row-inicio:row+inicio+1, column-inicio:column+inicio+1]
                        pass
    return img_copy    
```
