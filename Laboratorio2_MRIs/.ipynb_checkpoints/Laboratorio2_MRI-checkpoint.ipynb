## Binarazing The Image

Ingo ahead, I will use three methods to establish the threshold to binarize any image. This three methods are Otsu algorithm, random value and using an percentile value

## Otsu method

An Otsu method was described by Noboyuki Otsu. This method is based on disciminant criterion by separating the resultant classes in a gray levels. It uses the zeroth- and the first order cumulative moments of a gray-level histogram. [[1]](https://ieeexplore.ieee.org/stamp/stamp.jsp?tp=&arnumber=4310076).

```python
# Otsu's thresholding after Gaussian filtering
blur = cv2.GaussianBlur(img,(5,5),0)
ret3, img_otsu = cv2.threshold(blur,0,255,cv2.THRESH_OTSU)
print('The threshold found by otsu method was :',ret3)
#invert the image
img_otsu = cv2.bitwise_not(img_otsu)
# visualize 
plt.figure(figsize=(10,10))
plt.imshow(img_otsu, cmap = 'gray')
plt.title("Otsu's Thresholding")
plt.show()
```

## Random Manual Threshold

```python
thres_m = 0.3
# get copy of the image to modify
img_norm = np.copy(img)
# normalize the img
img_norm = (img_norm - img_norm.min())/(img_norm.max()-img_norm.min())
#create a img binarized 
bin_norm = np.zeros_like(img_norm)
bin_norm[(img_norm<=thres_m)]=1
# visualize 
plt.figure(figsize=(10,10))
plt.imshow(bin_norm, cmap='gray')
plt.title("Manual Thresholding")
plt.show()
```

## Percentile

Percentile means that the are a certain number o values that are under -th percentile. For instance, the 50 percentile is the median of the data, which the value 50-th is in the middle of the data [[2]]. 

```python
#get the threshol by value corresponding of percentil 80%
thres_p=np.percentile(img, 80)

img_per = np.zeros_like(img)
img_per[(img<=thres_p)]=1
# visualize 
plt.figure(figsize=(10,10))
plt.imshow(img_per, cmap='gray')
plt.title("Manual Thresholding")
plt.show()
```

## Gray scale

```python
#make zeros fully matrix to create the mask for each respective method
rem_pix_otsu = np.zeros_like(img, np.float32)
# replace the 1 to 0.75 gray scale level to ostu mask
rem_pix_otsu[(cv2.bitwise_not(img_otsu)==0)] = 0.75

# replace the 1 to 0.75 gray scale level to percentil method mask
rem_pix_per = np.zeros_like(img, np.float32)
rem_pix_per[(img_per==1)] = 0.75


# replace the 1 to 0.75 gray scale level to manual method methos mask
rem_pix_man = np.zeros_like(img, np.float32)
rem_pix_man[(bin_norm==1)] = 0.75
```

## Jaccard Score

```python
#laod the filea .mat
data_m= loadmat(os.path.join(wd,'jaccard.mat'))
#get the ground truth image
ground_t= data_m['GroundTruth']
#get the segmentation image
segmentation_data = data_m['Segmentation']
```

```python
#compare if two file has the same shape
segmentation_data.shape == ground_t.shape
```

```python
#convert the segmentation and ground truth to int 16 type data
seg = segmentation_data.astype(np.int16)
grd = ground_t.astype(np.int16)
np.unique(seg)
```

```python
def jaccard(x,y):
    x = np.asarray(x, np.bool) # Not necessary, if you keep your data
    y = np.asarray(y, np.bool) # in a boolean array already!
    # operation and (intersection) divided by operation or(union)
    return np.double(np.bitwise_and(x, y).sum()) / np.double(np.bitwise_or(x, y).sum())
```

```python
print("The Jaccard score was: ",jaccard(seg,grd))
```
