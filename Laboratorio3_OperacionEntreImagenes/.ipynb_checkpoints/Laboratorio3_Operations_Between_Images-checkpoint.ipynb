```python
# import all necesary libraries
import matplotlib.pyplot as plt
import cv2
import requests as rq
import numpy as np
import os, sys
import imageio
from scipy.io import loadmat
import pandas as pd
```

```python

def get_image(path, url):
    response = rq.get(url)
    
    with open(path, 'wb') as f:
        try:
            f.write(response.content)
        except:
            raise IOError('File not found. The root doesnÂ´t exist')
```

```python
def read_img(name):
    try:
        print('Visulazing the image stored in %s ...' %os.path.join(wd,name))
        # read the image with openCv
        img = cv2.imread(os.path.join(wd,name))
        # convert BGR to RGB order
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        # create a figure with 10x10
        plt.figure(figsize=(10,10))
        # show the image
        plt.imshow(img)
        plt.axis('off')
        return img
    except:
        raise NameError("The file {} doesn't exist".format(name))
```

```python
EminemSerious = read_img(path1)
```

```python
EminemSmile = read_img(path2)
```

```python
#create list to storage the combined image
images = []
for alpha in np.linspace(0,1,100):
    temp = np.zeros_like(EminemSerious)
    #apply I(x1, x2) = Î± x1 + (1 â Î±) x2.
    temp = alpha * EminemSerious + (1-alpha)*EminemSmile
    # cast to int variable
    temp.astype(int)
    images.append(temp)
```

```python
# another method (from scracth)
def elipse(img):
    #create a mask
    mask = np.zeros_like(img)  
    #get center coordinates of the ellipse
    a = img.shape[0]/2
    b = img.shape[1]/2  
    for row in range(img.shape[0]):
        for col in range(img.shape[1]): 
            if (np.abs(row-a)/a)**2 + (np.abs(col-b)/b)**2 < 1:
                #if it is inside (radius less than one)
                mask[row,col] = img[row,col]                
    return mask
```

## PrecisiÃ³n y Cobertura

```python
#load .m file
data_temp = loadmat(os.path.join('data','detection_results.mat'))
```

```python
#get annotations
annotations = data_temp['annotations']
#get annotations
predictions = data_temp['predictions'] 
#get scores
scores = data_temp['scores']
print('numero de datos...', len(scores))   
```

```python
# lists to store (precision, recall, fscore)
cober = []#storage the recall
prec = []
fscore = []
cont = 0
for e in np.linspace(0.01,1,100):#sort the data in ascending order
    TP = 0 # True positive
    FN = 1 # False negative
    FP = 0 #false positive
    for index in range(100):
        anotacion = annotations[index]
        deteccion = predictions[index]
        thre = scores[index]
        if thre>=e:
            #recreate the box bouding to ground true
            imagen_true = np.zeros((100,100),dtype= np.uint8)
            imagen_true[anotacion[1]:anotacion[1]+anotacion[3]-1,anotacion[0]:anotacion[0]+anotacion[2]-1]=1
            #recreate the box bouding to predicted box
            imagen_hat = np.zeros((100,100),dtype= np.uint8)
            imagen_hat[deteccion[1]:deteccion[1]+deteccion[3]-1,deteccion[0]:deteccion[0]+deteccion[2]-1]=1
            
            #Intersection Over Union (IOU) is measure based on Jaccard Index that evaluates the overlap between two bounding boxes.
            intersect = np.logical_and(imagen_true,imagen_hat)
            union = np.logical_or(imagen_true,imagen_hat)
            IOU = np.sum(intersect)/np.sum(union)
            if IOU<thre:
                FP+=1 #False Positive (FP): A wrong detection. Detection with IOU < threshold
            else:
                #True Positive (TP): A correct detection. Detection with IOU â¥ threshold
                TP+=1
    cobertura = TP/(TP+FN)#calculate recall
    presicion= 1 - TP/(TP+FP)
    Fmed=(2*presicion*cobertura)/(presicion+cobertura)#calculate fscore
    cober.append(cobertura)
    prec.append(presicion)
    fscore.append(Fmed)
    cont+=1
```

```python
plt.plot(cober,prec)
plt.title('Precision-Recall curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.tight_layout()
plt.savefig(os.path.join('img','curvePR.jpg'))
```

```python
plt.plot(np.linspace(0.01,1,100),fscore)
plt.title('F-Score-confidence curve')
plt.xlabel('Confidence')
plt.ylabel('F-Score')
plt.tight_layout()
plt.savefig(os.path.join('img','curveFscore.jpg'))
```

```python
print('El AP es de ...', np.mean(prec))
```

## Ecualizador

```python
get_image(os.path.join('img','histogram.jpg'),'https://ak2.picdn.net/shutterstock/videos/15390892/thumb/1.jpg')
img = read_img(os.path.join('img','histogram.jpg'))
```

```python
def my_histogram_equalizator(img, show_plot=True):
    #convert imgen to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    #flatten the matrix
    fl = gray.flatten()
    #get the histogram
    hist, bins = np.histogram(gray,256,[0,255])
    (height, width) = gray.shape
    #normalize the funciÃ³n de distribuciÃ³n acumulada (CDF)
    cdf = hist.cumsum()/(height*width)
    #scale and round toe ecualizer histogram
    temp = (np.round(255*(cdf))).astype('uint8')
    im2 = temp[fl]
    im3 = np.reshape(im2,gray.shape)
    if show_plot:
        plt.figure(figsize=(10,10))
        plt.subplot(1,2,1)
        plt.imshow(gray,cmap='gray')
        plt.title('Imagen bajo contraste (Original)')
        plt.axis('off')
        
        plt.subplot(1,2,2)
        plt.imshow(im3,cmap='gray')
        plt.title('Imagen alto contraste (Ecualizada)')
        plt.axis('off')
        
        plt.tight_layout()
        plt.savefig(os.path.join('img','eculiazada.jpg'))
```

```python
my_histogram_equalizator(img, show_plot=True)
```
