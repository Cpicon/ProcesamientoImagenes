```
import numpy as np
import skimage.io as io
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import requests
import cv2 
from google.colab import drive
import os
```

```
r = requests.get('https://www.marketingdirecto.com/wp-content/uploads/2015/09/steve-wozniak.jpg')
with open("Steve's.jpeg",'wb') as f:
  f.write(r.content)
```
